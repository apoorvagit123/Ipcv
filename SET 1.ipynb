```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
from scipy.ndimage import gaussian_filter

# Load your image
image_path = "img1.jpg"  # Replace with the correct name and extension
image = cv2.imread(image_path, cv2.IMREAD_GRAYSCALE)

if image is None:
    raise FileNotFoundError(f"Image {image_path} not found. Please check the file name and path.")

# Apply Gaussian smoothing
smoothed_image = gaussian_filter(image, sigma=1)

# Compute gradients using the Sobel operator
sobel_x = cv2.Sobel(smoothed_image, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(smoothed_image, cv2.CV_64F, 0, 1, ksize=3)

# Compute edge magnitude
magnitude = np.sqrt(sobel_x**2 + sobel_y**2)

# Plot results
plt.figure(figsize=(12, 6))
plt.subplot(1, 4, 1), plt.imshow(image, cmap='gray'), plt.title('Original Image')
plt.subplot(1, 4, 2), plt.imshow(smoothed_image, cmap='gray'), plt.title('Smoothed Image')
plt.subplot(1, 4, 3), plt.imshow(np.abs(sobel_x), cmap='gray'), plt.title('Sobel X')
plt.subplot(1, 4, 4), plt.imshow(magnitude, cmap='gray'), plt.title('Edge Magnitude')
plt.tight_layout()
plt.show()
```
